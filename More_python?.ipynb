Now we are ready to see some fundamentals on python... we are going to be using a simple data set
1. Variables: Store data values. In data science, we often use them for datasets, columns, metrics, etc.
2. Conditionals (if, elif, else): Used to control the flow of logic ‚Äî essential in data cleaning, filtering, and decision making.
3. Lists (or Arrays): A list holds multiple values. Great for storing features, column names, predictions, etc.
4. Loops: Loops let you repeat operations ‚Äî useful for iterating through data, features, rows, etc.
5. Functions: Functions let you reuse code ‚Äî essential in pipelines, preprocessing, metrics, and models.



```python
import pandas as pd  # Import the pandas library

# üß™ Create a simple example dataset using a dictionary
data = {
    "name": ["Alice", "Bob", "Charlie", "Diana"],
    "age": [29, 35, 22, 31],
    "salary": [48000, 61000, 39000, 72000],
    "education": ["Bachelor", "Master", "Bachelor", "PhD"]
}

# Convert the dictionary to a DataFrame (table-like structure)
df = pd.DataFrame(data)

# üü¢ VARIABLES: Store useful stats like mean
mean_age = df["age"].mean()       # Average age of all people
#pd.DataFrame(data["age"]).mean()
mean_salary = df["salary"].mean() # Average salary

print(f"Mean Age: {mean_age}") #print("Mean Age: {}".format(mean_age)) print("Mean Age: %.2f" % mean_age)
print(f"Mean Salary: {mean_salary}")

# üîµ CONDITIONALS: Make decisions based on data
if mean_salary > 50000:
    print("Above average salary")
elif mean_salary == 50000:
    print("Exactly average salary")
else:
    print("Below average salary")

# üü° LISTS: Use Python lists or list-like structures
columns = list(df.columns)  # Get column names as a list
print("Dataset Columns:", columns)

# üîÅ LOOPS: Loop over column names and print data types
for col in df.columns:
    print(f"{col} is of type {df[col].dtype}")

# üîß FUNCTIONS: Define reusable logic
#Statistics
# Function to normalize a column (mean 0, std 1)
def normalize(column):
    return (column - column.mean()) / column.std()

# Apply normalization to the salary column
df["normalized_salary"] = normalize(df["salary"])

# Function to classify salary into 'High' or 'Low'
def classify_income(salary):
    return "High" if salary > 60000 else "Low"

# Apply classification to each salary value
df["income_group"] = df["salary"].apply(classify_income)

# üñ®Ô∏è Final dataset preview
print("\nUpdated DataFrame:")
print(df)
```

**NumPy**

NumPy means Numerical Python. It is an open-source library used to perform mathematical tasks with very high efficiency. In addition, it introduces data structures, such as multidimensional arrays, which can be operated on at a high level, without getting too much into the details.

Specifically, the keys to this library are:

* Multidimensional arrays: This library provides an object called ndarray, which allows you to store and manipulate large data sets efficiently. Arrays can have any number of dimensions.
* Vectorized operations: NumPy allows performing mathematical operations on complete arrays without the need for explicit loops in the code, which makes it very fast and efficient.
* Mathematical functions: NumPy provides a wide range of mathematical functions for working with arrays, including trigonometric functions, statistics, and linear algebra, among others.
* Efficiency: It is much faster than the same functionality implemented directly on native Python. It is also very flexible in terms of accessing and manipulating individual elements or subsets of arrays.

NumPy is a fundamental library for Machine Learning and data science in Python. It provides a wide range of tools and functions to work efficiently with numerical data in the form of arrays and matrices.
